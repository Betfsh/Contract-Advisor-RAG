```python
import json

def extract_json(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        return data
    except Exception as e:
        print(f"An error occurred while extracting JSON data: {e}")
        return None

# Example usage:
#file_path = "../data/prepare_policy_data/data/qa_with_context/dev.json"
file_path = "../data/qa_with_context/dev.json"

json_data = extract_json(file_path)
if json_data:
    print("JSON data extracted successfully!")
    # Do further processing with the extracted JSON data
    # For example, print the content:
    print(json_data)
else:
    print("Failed to extract JSON data.")
```
